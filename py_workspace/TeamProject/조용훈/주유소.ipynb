### 망한 주유소 찾기

```python
names= seoul_oil.pivot_table(index=["월","상호"]).index
names
```

```python
names=names.to_frame(index=False)
```

###### max와 min을 월에 적용해 시작일과 끝 일을 찾기

```python
names
end=names.groupby("상호").max()
end.rename(columns = {'월': 'end'}, inplace = True)

start=names.groupby("상호").min()
start.rename(columns = {'월': 'start'}, inplace = True)
```

###### 두개를 합처서 시작일과 끝 일을 담고있는 데이터 프레임 생성

```python
start_end= pd.concat([start,end],axis=1)
```

###### 망한 주유소와 새로 개업한 주유소 검색

```python
closed=start_end[start_end["end"]!=12]
closed
```

```python
opened=start_end[start_end["start"]!=1]
opened
```

###### 월별 망한 주유소 수와 새로 개업한 주유소 수 그리기

```python
closed.pivot_table(index="end",aggfunc="count").plot(kind="bar",figsize=(10,5))
plt.xlabel("월")
plt.ylabel("망한 주유소 수")
plt.plot()
```

```python
opened.pivot_table(index="start",aggfunc="count").plot(kind="bar",figsize=(10,5))
plt.xlabel("월")
plt.ylabel("개업한 주유소 수")
plt.gca().get_legend().remove()
plt.plot()
```

### 구별 가격 비교, 지역별 소득과 휘발유 값의 상관관계

```python
area=seoul_oil.pivot_table(index="지역",values="휘발유",aggfunc="mean")
```

```python
income=pd.read_csv("../data/2019년도 4월기준 14_18년 시군구별 평균소득월액..csv",encoding="cp949")
income
```

```python
income.rename(columns={"시군구":"지역"},inplace=True)
seoul_income=income[income["지역"].str.contains("서울")]
seoul_income["지역"]=seoul_income["지역"].str.replace("서울특별시","서울")

# seoul_income["지역"].isin(area.index)
```

```python
seoul_income.set_index(seoul_income["지역"],inplace=True)
seoul_income.drop(axis=1,columns=["지역","2014년","2015년","2016년","2017년"],inplace=True)
```

```python
price_income=pd.concat([area,seoul_income],axis=1)
```

```python
price_income.plot(kind= "bar", y="휘발유")
```

```python
graph1=price_income
plt.figure(figsize=(20,10))
ax=graph1.plot(kind="line",y="휘발유",color="blue")
ax2=graph1.plot(kind="line",y="2018년",secondary_y=True,color="Red",ax=ax)

ax.set_ylabel('휘발유가격')
ax2.set_ylabel('소득')
# plt.tight_layout()
plt.plot()
plt.show()
```
