## 데이터불러오기1

```python
# seoul_oil1 = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv',  encoding='utf-8')
seoul_oil1 = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv',  encoding='cp949')
seoul_oil1
```

```python
# seoul_oil1 = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv',  encoding='utf-8')
seoul_oil2 = pd.read_csv('../data/2019년 하반기 주유소 판매가격.csv',  encoding='cp949')
seoul_oil2
```

```python
seoul_oil = pd.concat([seoul_oil1,seoul_oil2])
seoul_oil
```

```python
seoul_oil['기간'] = pd.to_datetime(seoul_oil['기간'], format='%Y%m%d', errors='raise')
seoul_oil.info()
```

```python
seoul_oil['월'] = seoul_oil['기간'].dt.month
seoul_oil
```

```python
#고급휘발유가 있는 주유소들
high_oil = seoul_oil[seoul_oil['고급휘발유'] != 0]
high_oil
```

```python
#고급휘발유를 취급하는 주유소들... 짱많음
high_oil['상호'].unique()
```

```python
# 고급휘발유 취급하는 주유소들의 고급휘발유와 휘발유 평균 가격
high_oil.pivot_table(index=['지역','상호'], values=['고급휘발유','휘발유'], columns='월')
high_oil.pivot_table(index=['지역','상호'], values=['고급휘발유','휘발유'])
```

```python
#고급휘발유가 없는 주유소들
low_oil = seoul_oil[seoul_oil['고급휘발유'] == 0]
low_oil
```

```python
# 고급휘발유 없는 주유소들의 휘발유 평균 가격
low_oil.pivot_table(index=['지역','상호'], values='휘발유', columns='월')
low_oil.pivot_table(index=['지역','상호'], values='휘발유')
```

```python
#고급휘발유 취급하는 데 휘발유값 월별 평균
high_oil.pivot_table(index=['지역','상호'], values='휘발유')
high_oil.pivot_table(index='지역', values='휘발유')
high = high_oil.pivot_table(index='월', values='휘발유')
# high = high_oil.groupby('월')[['휘발유']].mean()
high
```

```python
#고급휘발유 없는데 휘발유값 월별 평균
low_oil.pivot_table(index=['지역','상호'], values='휘발유')
low_oil.pivot_table(index='지역', values='휘발유')
low = low_oil.pivot_table(index='월', values='휘발유')
# low = low_oil.groupby('월')[['휘발유']].mean()
low
```

```python
#월별로 고급휘발유 있는곳과 없는곳의 휘발유 가격 비교
plt.figure(figsize=(15,5))

plt.plot(high, 'y', lw=4, label='고급휘발유 취급')    # lw는 line width
plt.plot(low, 'g', lw=4, label='일반휘발유만 취급')
plt.grid()
plt.legend()
plt.xlabel('월')
plt.ylabel('휘발유가격')
plt.title('고급휘발유 있는 주유소는 휘발유값이 더 비쌀까?')
plt.show()
```

```python
#고급휘발유 있는곳과 없는곳의 지역별 휘발유값 
#보기쉽게 컬럼명 바꾸기
high_oil_area = high_oil.pivot_table(index='지역', values='휘발유')
high_oil_area.rename(columns={"휘발유":"고급휘발유 취급"},inplace=True)    #고급휘발유 있는곳
high_oil_area
low_oil_area = low_oil.pivot_table(index='지역', values='휘발유')
low_oil_area
low_oil_area.rename(columns={"휘발유":"고급휘발유 무취급"},inplace=True)    #고급휘발유 없는곳
all = pd.concat([high_oil_area,low_oil_area],axis=1)
all.sort_values('고급휘발유 취급', ascending=False, inplace=True)
all
```

```python
#고급휘발유 있는곳의 지역별 휘발유값 
high_oil_area = high_oil.pivot_table(index='지역', values='휘발유')
# high_oil_area.rename(columns={"휘발유":"휘발유1"},inplace=True)    #고급휘발유 있는곳
high_oil_area
```

```python
#고급휘발유 없는곳의 지역별 휘발유값 
low_oil_area = low_oil.pivot_table(index='지역', values='휘발유')
low_oil_area
```
