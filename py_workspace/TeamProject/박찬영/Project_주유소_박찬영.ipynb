```python
# seoul_oil1 = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv',  encoding='utf-8')
seoul_oil1 = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv',  encoding='cp949')
seoul_oil1
```

```python
# seoul_oil1 = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv',  encoding='utf-8')
seoul_oil2 = pd.read_csv('../data/2019년 하반기 주유소 판매가격.csv',  encoding='cp949')
seoul_oil2
```

```python
seoul_oil = pd.concat([seoul_oil1,seoul_oil2])
seoul_oil
```

```python
seoul_oil['기간'] = pd.to_datetime(seoul_oil['기간'], format='%Y%m%d', errors='raise')
seoul_oil.info()
```

```python
seoul_oil['월'] = seoul_oil['기간'].dt.month
seoul_oil
```

```python
brand_self_check= seoul_oil.pivot_table(['휘발유','경유'],index='상표',columns='셀프여부')
brand_self_check
```

```python
brand_self_check.plot(kind='bar')
plt.show()
```

```python
self_oil = seoul_oil[seoul_oil['셀프여부'].isin(['셀프'])] # 셀프 여부 컬럼에서 '셀프'가 있는 테이블을 뽑아냄 
self_oil
```

```python
test=self_oil[['상표']]
test.drop_duplicates()
```

```python
self_oil2= self_oil[['지역','상호','상표']] # 특정 값만 추출
self_oil2.drop_duplicates(inplace=True) # 중복값 제거
self_oil2.reset_index(inplace=True) # 인덱스 재 정렬
self_pivot= self_oil2.pivot_table('상호',index='지역',columns='상표',aggfunc='count') # 피봇 테이블 생성
self_pivot=self_pivot.fillna(0) # NaN 값 0으로 변경
self_pivot= self_pivot.astype(int) # 데이타 타입 int로 변경
sns.heatmap(self_pivot,annot=True,fmt='d') 
plt.show()
```
