### 1.DataLoading and Data PreProcessing

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('../data/movies.csv')
ratings = pd.read_csv('../data/ratings_1.csv')
print(movies.shape)
print(ratings.shape)
```

```python
movies.head()
```

```python
'''
ratings는 사용자별로 영화에 대한 평점을 매긴 데이타셋
평점은 최소 0.5 ~ 최대 5 사이이며 0.5단위로 평점이 부여되어 있다.
timestamp 컬럼은 1도 필요없다
'''
ratings.head()
```

```python
# 필요한 컬럼만으로 다시 Dataframe을 생성
ratings = ratings[['userId','movieId', 'rating']]
ratings

# 사용자를 row, 영화정보를 컬럼으로 구성한 피봇 생성
ratings_matrix = ratings.pivot_table('rating', index='userId', columns='movieId')
ratings_matrix
```

```python
'''
영화 title이 필요하다.... id대신에
title컬럼은 movies에 있다

title컬럼을 얻기위해서는 movies와 조인해야한다.
'''

rating_movies = pd.merge(ratings, movies, on='movieId')
rating_movies
```

```python
# 1. pivot_table... index=userid, columns=title
ratings_matrix = rating_movies.pivot_table(index='userId', columns='title', values='rating')
ratings_matrix
```

```python
#2. NaN값을 모두 0으로 변환
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix
```

### 2. 영화와 영화들 간의 유사도 산출 - Cosine Similarity
    사용자와 변환된 영화 평점 데이타 행렬 데이타셋을 이용해서 영화간 유사도를 측정해보자
    코사인 유사도는 기준이 되는 행과 다른 행을 비교해서 유사도를 특정하는 함수이다.
    지금과 같은 행렬구조는 userId가 기준인 행 레벨 데이타이기 때문에 사용자 간의 유사도가 만들어진다.
    이것을 사용자가 아닌 영화와 영화들간의 유사도를 만들려면 어떻게 해야할까?

    참고로 코사인 유사도 라이브러리는 사이킷런의 cosine_similarity() 함수를 사용한다

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)
item_sim_df = pd.DataFrame(item_sim, index=ratings_matrix.columns, columns=ratings_matrix.columns)
item_sim_df
```

```python
# item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)
item_sim_df["Godfather, The (1972)"].sort_values(ascending=False).head(6)
```

```python
item_sim_df['Inception (2010)'].sort_values(ascending=False)[1:6]
```
