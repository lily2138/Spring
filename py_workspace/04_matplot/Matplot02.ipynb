### 히스토그램그리기

```python
'''
hist() 함수 사용
히스토그램은 자료의 도수분포상태를 직사각형 막대그래프로 나타낸것
데이타의 빈도(도수)에 따라서 높이가 결정된다.
plot() 함수를 이용해서 직선 혹은 꺾은선을 
scatter() 함수를 이용해서 산포도(산점도)를 그렸듯이
hist() 함수를 이용해서 막대그래프를 그려보자
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import seaborn as sns
```

```python
plt.hist([1,1,2,3,4,5,6,6,6,7,8,10])
```

```python
dice = []
for i in range(5):
    dice.append(np.random.randint(1,7))
print(dice)
```

```python
'''
bins가 작으면 막대그래프가 뭉뚱그려진다.
bins가 너무 큰값이면 이빨처럼 촘촘해진다.
히스토그램을 그릴때는 bins값을 잘 지정하는 것이 중요
기본값도 적당하다.

bins
해당 막대의 영역을 얼마나 채울건지를 정하는 것
'''
plt.hist(dice, bins=15)
plt.show()
```

```python
'''
주사위를 던지는 횟수를 늘릴수록 특정한 숫자가 나오는 횟수가 전체의 1/6에 가까워진다...
'큰수의 법칙'을 히스토그램으로 확인
'''
for i in range(1000000):
    dice.append(np.random.randint(1,7))
plt.hist(dice)
plt.show()
```

```python
n, bins, patches = plt.hist(x, bins=50, density=1, facecolor='g', alpha=0.6)   #alpha는 투명도 0~1
plt.xlabel('Smarts')
plt.ylabel('Probablity')
plt.title('Histogram of info')
plt.grid()
plt.text(60, 0.025, '$mu=100, \sigma=15$')
plt.axis([40, 160,0, 0.03])      # 가로세로 범위 정해서 보여주기 
plt.show()
```

### BoxPlot이상치 값 제거하기

```python
'''
boxplot이상치값 제거하는 방법
1) 이상치값 제거
2) 이상치값 변환 - - NaN많이 사용함.
'''
```

```python
z = tips.boxplot(column=['tip'], by=['day', 'sex' ])
plt.xticks(rotation=45)
```

```python
#이상치값 제거 함수
def removeOutlier(x, column):
    #Q1, Q3 구하기
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    
    #iqt(IQR 범위를 벗어난 값...4분위 범위를 벗어난 값)
    iqt = 1.5*(q3-q1)
    
    #원래 데이타 복제
    y=x
    
    y['tip'][(tips['tip']>(q3+iqt)) | (tips['tip']<(q1-iqt))] = None
    y['tip']
    
    return(y)
```

```python
removeOutlier(tips,'tip').boxplot(column='tip', by=['day', 'sex'])
plt.xticks(rotation=45)
plt.show()
```

### HeatMap

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot_table(index='month', columns='year', values='passengers')
flights
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights)      #디폴트값이 안보인다
plt.show()
```

```python
'''
x축은 year
y축은 month가 된다.

값이 점점 커지는 쪽.... 점점 환해지는 쪽....
1960년 7월.... 휴가철

'''
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d')      #디폴트값이 안보인다
plt.show()
```
